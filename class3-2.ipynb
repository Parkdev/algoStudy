```python
#18870 좌표압축
# 수의 범위가 매우 큰 상태에서 수의 값과 상관없이 숫자 간의 대소관계만 알면 될때 이용하는 알고리즘
import sys

n = int(sys.stdin.readline())
nums = list(map(int,sys.stdin.readline().split()))
coord = sorted(list(set(nums)))

def binary_search(target, data):
  start = 0
  end = len(data)-1

  while start <= end:
    mid = (start+end)//2
    if data[mid] == target:
      return mid
    if data[mid] >= target:
      end = mid -1
    else:
      start = mid + 1
  return -1

for i in nums:
  print(binary_search(i,coord), end=' ')
```
